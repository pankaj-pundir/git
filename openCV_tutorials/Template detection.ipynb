```python
## finding sunflower
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
shape=cv2.imread('sunflower.jpg',cv2.IMREAD_COLOR)
waldosPic=cv2.imread('sunflowersPic.jpg',0)
plt.imshow(shape)
```

```python
gray_shape=cv2.cvtColor(shape,cv2.COLOR_RGB2GRAY)
res=cv2.matchTemplate(gray_shape,waldosPic,cv2.TM_CCOEFF)
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
# cv2.circle(shape,max_loc,5,(0,0,0),4)
# cv2.circle(shape,min_loc,5,(0,0,0),4)
cv2.rectangle(shape,max_loc,(max_loc[0]+waldosPic.shape[0],max_loc[1]+waldosPic.shape[1]),(2,0,255),2)
cv2.putText(shape,'Found it',(int(shape.shape[0]/2-20),int(shape.shape[1]/20)),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,255),2)
print(min_val)
plt.imshow(shape)
```
