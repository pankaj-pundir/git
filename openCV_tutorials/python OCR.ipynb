```python
def remove_text(img):
    boxes=pyt.image_to_boxes(quick_fox)
    for b in boxes.splitlines():
        b = b.split(' ')
        charLen_x=int((int(b[3])-int(b[1]))*0.5)
        charLen_y=int((int(b[4])-int(b[2]))*0.5)
        img = cv2.rectangle(quick_fox, (int(b[1])-charLen_x,h - int(b[2])+charLen_y), (int(b[3])+charLen_x, h - int(b[4])-charLen_y), (255, 255, 0), 1)
    return img
```
