```python
## FAce and Eye recognition

#     haarcascade_frontalface_alt.xml
#     haarcascade_eye.xml
#     haarcascade_smile.xml
#     haarcascade_frontalface_alt_tree.xml

face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
eye_classifier=cv2.CascadeClassifier('haarcascade_eye.xml')

def face_eye_detection(img):
    #img=cv2.imread('shrey_face.jpeg')
    gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    faces=face_classifier.detectMultiScale(gray,1.1,5)
    if faces == ():
        return img
    else:
        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
            cropped_image=gray[y:y+h,x:x+w]
            eyes=eye_classifier.detectMultiScale(cropped_image,1.1,10)

            if eyes == ():
                return img
            else:
                for (ex,ey,ew,eh) in eyes:
                    cv2.rectangle(img,(x+ex,y+ey),(x+ex+ew,y+ey+eh),(0,255,0),4)
    return img
   

```

```python
img=cv2.imread('me_image.jpg')
plt.imshow(face_eye_detection(img))
```

```python
## Video Live face and eye detection
cap=cv2.VideoCapture(0)
fourcc=cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('output.',fourcc,20,(640,480))
while True:
    ret,frame=cap.read()
    cv2.imshow("skecth",face_eye_detection(frame))
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        cap.release()
        cv2.destroyAllWindows()
```
